```python
# Cell 1 - imports and filename
import numpy as np
import matplotlib.pyplot as plt
import math

fname = "scrambled.npy"   # <-- change this if your file has a different name
```

```python
# Cell 2 - load .npy and inspect
arr = np.load(fname)
print("arr.shape:", arr.shape)
print("total elements:", arr.size)
```

```python
# Cell 3 - compute sqrt and reshape (assumes flattened 1D or reshape-able)
total = arr.size
n = int(math.isqrt(total))
mat = arr.reshape(n, n)

print("mat.shape:", mat.shape)
```

```python
# Cell 4 - show raw reshaped image
plt.figure(figsize=(5,5))
plt.imshow(mat, cmap='gray')
plt.title("raw reshaped")
plt.axis('off')
plt.show()
```

```python
# Cell 5 - Correct orientation
final_mat = np.flipud(mat)   # replace with the correct one

plt.imshow(final_mat, cmap='gray')
plt.title("Final corrected")
plt.axis('off')
plt.show()

plt.imsave("decoded_image.png", final_mat, cmap='gray')
print("Saved decoded_image.png")
```
